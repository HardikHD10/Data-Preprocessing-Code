```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from skimage import io
import os
from PIL import Image
import random
```

```python
# Path to Datasets
# Training Path
train_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/training/caries/"
train_no_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/training/without_caries/"

# Testing Path 
test_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/test/caries/"
test_no_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/test/no-caries/"
```

```python
# Making Directory of Images
# Training Images
train_caries_image = os.listdir(train_caries_path)
train_no_caries_image = os.listdir(train_no_caries_path)


#Testing Images
test_caries_image = os.listdir(test_caries_path)
test_no_caries_image = os.listdir(test_no_caries_path)
```

```python
# Printing Numbers of Images in each Dataset
print("Number of Caries Images in Training = ",len(train_caries_image))
print("Number of No Caries Images in Training = ",len(train_no_caries_image))


print('\n')
print("Number of Caries Images in Testing = ",len(test_caries_image))
print("Number of No Caries Images in Training = ",len(test_no_caries_image))
```

```python
valid_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/validation/caries/"
valid_no_caries_path = "/content/drive/MyDrive/teeth_dataset/teeth_dataset/validation/no-caries/"
```

```python
#Validation Images
valid_caries_image = os.listdir(valid_caries_path)
valid_no_caries_image = os.listdir(valid_no_caries_path)
```

```python
print('\n')
print("Number of Caries Images in Validation = ",len(valid_caries_image))
print("Number of No Caries Images in Validation = ",len(valid_no_caries_image))
```

# **Part - 2**

```python
datagen = ImageDataGenerator(
    rotation_range = 45,
    width_shift_range = 0.3,
    height_shift_range = 0.3,
    brightness_range = (0.1,0.9),
    horizontal_flip = True,
    vertical_flip = True,
    fill_mode = "nearest"
)
```

# **Part - 3**

Test : caries

```python
SIZE = 150
dataset_1 = []

for i, image_name in enumerate(test_caries_image):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'):
        image = io.imread(test_caries_path + '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset_1.append(np.array(image))

x = np.array(dataset_1)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/teeth_dataset/Augmented/teeth_dataset/test/caries', 
                          save_prefix='aug', 
                          save_format='png'):
    i += 1
    if i > 20:
        break
```

Test  : Not - caries

```python
SIZE = 150
dataset_2 = []

for i, image_name in enumerate(test_no_caries_image):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'):
        image = io.imread(test_no_caries_path + '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset_2.append(np.array(image))

y = np.array(dataset_2)
```

```python
i = 0
for batch in datagen.flow(y,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/teeth_dataset/Augmented/teeth_dataset/test/no-caries', 
                          save_prefix='aug', 
                          save_format='png'):
    i += 1
    if i > 20:
        break
```
