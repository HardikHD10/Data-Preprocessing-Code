```python
from PIL import Image
import os

def resize_lossless_tiff(input_image, output_image, scale_factor=0.316):
    image = Image.open(input_image).convert('RGB')
    width, height = image.size
    new_width, new_height = int(180 * scale_factor), int(60 * scale_factor)
    
    resized_image = image.resize((new_width, new_height), resample=Image.LANCZOS)
    resized_image.save(output_image, compression='tiff_lzw')
```

```python
import os
input_directory_0 = "/content/drive/MyDrive/Cache/AI - Sem 6/HICD/main/Training/NORMAL"
output_directory_0 = "/content/drive/MyDrive/Cache/AI - Sem 6/Histopathologic Oral Cancer Detection/Training/NORMAL"
if not os.path.exists(output_directory_0):
    os.makedirs(output_directory_0)

for filename in os.listdir(input_directory_0):
    if filename.lower().endswith(('.tif', '.tiff')):
        input_image = os.path.join(input_directory_0, filename)
        output_image = os.path.join(output_directory_0, filename)
        resize_lossless_tiff(input_image, output_image, scale_factor=0.4)
```
