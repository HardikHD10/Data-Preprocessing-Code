```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from skimage import io
import os
from PIL import Image
import random
```

```python
# Path to Datasets
# Training Path
Bengin = "/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Bengin cases"
Malignant = "/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Malignant cases"
Normal = "/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Normal cases"
```

```python
#training Data
image_Begin = os.listdir(Bengin)
image_Malignant = os.listdir(Malignant)
image_Normal = os.listdir(Normal)
```

```python
print("Number of Benigin = ",len(image_Begin))
print("Number of Malignant Cases= ",len(image_Malignant))
print("Number of Normal Cases= ",len(image_Normal))
```

```python
datagen = ImageDataGenerator(
    brightness_range = (0.1,0.9),
    horizontal_flip = True,
    fill_mode = "nearest"
)
```

```python
SIZE = 128
dataset_1 = []

for i, image_name in enumerate(image_Begin):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'):
        image = io.imread(Bengin + '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset_1.append(np.array(image))

x = np.array(dataset_1)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Augumented/test_benign_aug', 
                          save_prefix='aug', 
                          save_format='png'):
    i += 1
    if i > 50:
        break
```

```python
SIZE = 128
dataset_2 = []

for i, image_name in enumerate(image_Malignant):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'):
        image = io.imread( Malignant+ '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset_2.append(np.array(image))

y = np.array(dataset_2)
```

```python
i = 0
for batch in datagen.flow(y,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Augumented/test_malignant_aug', 
                          save_prefix='aug', 
                          save_format='png'):
    i += 1
    if i > 50:
        break
```

```python
SIZE = 128
dataset_3 = []

for i, image_name in enumerate(image_Normal):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'):
        image = io.imread( Normal+ '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((SIZE,SIZE))
        dataset_3.append(np.array(image))

z = np.array(dataset_3)
```
