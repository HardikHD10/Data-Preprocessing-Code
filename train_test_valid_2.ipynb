```python
import os
import shutil
import numpy as np
```

```python
dataset_path = "/content/drive/MyDrive/AI - Sem 6/HICD/main"
```

```python
folders = ['NORMAL', 'OSCC']
```

```python
train_ratio = 0.7
test_ratio = 0.15
val_ratio = 0.15
```

```python
# Creating the final folder structure
final_folders = {
    'Training': os.path.join(dataset_path, 'Training'),
    'Testing': os.path.join(dataset_path, 'Testing'),
    'Validation': os.path.join(dataset_path, 'Validation')
}

for folder_name, folder_path in final_folders.items():
    os.makedirs(folder_path, exist_ok=True)
    for class_folder in folders:
        os.makedirs(os.path.join(folder_path, class_folder), exist_ok=True)
```

```python
# Splitting and copying the images
for class_folder in folders:
    src_path = os.path.join(dataset_path, class_folder)
    img_files =  os.listdir(src_path)
    np.random.shuffle(img_files)

    n_total_files = len(img_files)
    n_train = int(n_total_files * train_ratio)
    n_test = int(n_total_files * test_ratio)

    for i, img_file in enumerate(img_files):
        src_file_path = os.path.join(src_path, img_file)
        if i < n_train:
            dest_folder = final_folders['Training']
        elif i < n_train + n_test:
            dest_folder = final_folders['Testing']
        else:
            dest_folder = final_folders['Validation']
        
        dest_file_path = os.path.join(dest_folder, class_folder, img_file)
        shutil.copy2(src_file_path, dest_file_path)

print("Dataset split and copied successfully.")
```

##Resizing

```python
import cv2
import os

# Set the input and output directories
input_dir0 = '/content/drive/MyDrive/AI - Sem 6/HICD/main/Testing/NORMAL'
output_dir0 = '/content/drive/MyDrive/AI - Sem 6/HICD/Resized_main/Testing/NORMAL'

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir0):
    os.makedirs(output_dir0)

# Set the new image size
new_width = 160
new_height = 80

# Set the scaling factor
scale_factor = 0.4

# Loop through all the files in the input directory
for filename in os.listdir(input_dir0):
    # Read the image
    img = cv2.imread(os.path.join(input_dir0, filename))
    
    # Resize the image
    img = cv2.resize(img, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)
    img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_AREA)
    
    # Save the resized image
    cv2.imwrite(os.path.join(output_dir0, filename), img)

print('Images resized successfully!')
```
