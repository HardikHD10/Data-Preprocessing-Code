```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from skimage import io
import os
from PIL import Image
import random
```

```python
datagen = ImageDataGenerator(
    rotation_range = 60,
    horizontal_flip = True,
    vertical_flip = True,
    width_shift_range=0.3,
    height_shift_range=0.3,
    fill_mode='wrap',
)
```

# **Training**

```python
train_MCL_path = "/content/drive/MyDrive/AI - Sem 6/Malignant Lymphoma Classification/Dataset/dataset_1/Training/MCL"
train_FL_path = "/content/drive/MyDrive/AI - Sem 6/Malignant Lymphoma Classification/Dataset/dataset_1/Training/FL"
train_CLL_path = "/content/drive/MyDrive/AI - Sem 6/Malignant Lymphoma Classification/Dataset/dataset_1/Training/CLL"
```

```python
train_MCL_image = os.listdir(train_MCL_path)
train_FL_image = os.listdir(train_FL_path)
train_CLL_image = os.listdir(train_CLL_path)
```

```python
print('\n')
print("Number of MCL Images in Training = ",len(train_MCL_image))
print("Number of FL Images in Training = ",len(train_FL_image))
print("Number of CLL Images in Training = ",len(train_CLL_image))
```

```python
dataset_0 = []

for i, image_name in enumerate(train_MCL_image):
    if (image_name.split('.')[1] == 'jpg' or image_name.split('.')[1] == 'jpeg' or image_name.split('.')[1] == 'png'or image_name.split('.')[1] == 'tif'):
        image = io.imread(train_MCL_path + '/' + image_name)
        image = Image.fromarray(image, 'RGB')
        image = image.resize((224,224))
        dataset_0.append(np.array(image))

x = np.array(dataset_0)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/AI - Sem 6/Malignant Lymphoma Classification/Dataset/Augmented_Dataset/Training/MCL', 
                          save_prefix='aug', 
                          save_format='tif'):
    i += 1
    if i > 20:
        break
```

```python
import os
import random

# Set the path to the directory containing the images
image_dir = '/content/drive/MyDrive/AI - Sem 6/Malignant Lymphoma Classification/Dataset/Augmented_Dataset/Validation/MCL'

# Get a list of all the image filenames in the directory
image_filenames = os.listdir(image_dir)

# Shuffle the list of filenames randomly
random.shuffle(image_filenames)

# Select the first 101 filenames from the shuffled list
images_to_delete = image_filenames[:10]

# Loop over the selected filenames and delete the corresponding files
for filename in images_to_delete:
    file_path = os.path.join(image_dir, filename)
    os.remove(file_path)
    print('Deleted:', filename)
```
