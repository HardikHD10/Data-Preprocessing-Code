```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Read the dataset
# Replace 'input_file.csv' with the path to your CSV file
input_file = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 3 Disease diagnosis based on symptoms /combined_dataset.csv'
df = pd.read_csv(input_file)

# Initialize empty DataFrames for training and testing sets
train_df = pd.DataFrame()
test_df = pd.DataFrame()

# Extract 20% of samples for each disease (prognosis)
unique_prognoses = df['prognosis'].unique()
for prognosis in unique_prognoses:
    prognosis_df = df[df['prognosis'] == prognosis]
    
    prognosis_train, prognosis_test = train_test_split(prognosis_df, test_size=0.2)
    
    train_df = pd.concat([train_df, prognosis_train], ignore_index=True)
    test_df = pd.concat([test_df, prognosis_test], ignore_index=True)

# Save the training and testing sets to CSV files in Google Drive
train_output_path = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 3 Disease diagnosis based on symptoms /train_dataset.csv'
test_output_path = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 3 Disease diagnosis based on symptoms /test_dataset.csv'

train_df.to_csv(train_output_path, index=False)
test_df.to_csv(test_output_path, index=False)
```

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split

# Install imbalanced-learn library if needed
!pip install imbalanced-learn

# Read the dataset
input_file = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 3 Disease diagnosis based on symptoms /Testing.csv'
df = pd.read_csv(input_file)

# Separate the features and target variable
X = df.drop('prognosis', axis=1)
y = df['prognosis']

# Perform random oversampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Combine the resampled features and target variable into a single DataFrame
oversampled_df = pd.concat([X_resampled, y_resampled], axis=1)

# Save the oversampled dataset to a CSV file in Google Drive
oversampled_output_path = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 3 Disease diagnosis based on symptoms /Code/oversampled_dataset.csv'
oversampled_df.to_csv(oversampled_output_path, index=False)

print(f"Oversampled dataset saved at: {oversampled_output_path}")
```
