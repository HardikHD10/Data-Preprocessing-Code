```python
import os
import random
import shutil
```

```python
# Set the source directory containing all the images
source_dir = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Augumented/test_benign_aug'

# Set the target directories for train, validation, and test data
train_dir = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Final_Dataset/Data/Training/benign'
val_dir = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Final_Dataset/Data/Validation/benign'
test_dir = '/content/drive/MyDrive/Big Data - Sem 6/Problem Statement 1 The IQ-OTHNCCD lung cancer dataset/Final_Dataset/Data/Testing/benign'

# Set the percentage split for train, validation, and test data
train_split = 0.7
val_split = 0.2
test_split = 0.1

# Get the list of all image files in the source directory
image_files = os.listdir(source_dir)

# Shuffle the image files randomly
random.shuffle(image_files)

# Calculate the number of images for each split
train_size = int(train_split * len(image_files))
val_size = int(val_split * len(image_files))
test_size = len(image_files) - train_size - val_size

# Split the image files into train, validation, and test sets
train_images = image_files[:train_size]
val_images = image_files[train_size:train_size+val_size]
test_images = image_files[train_size+val_size:]

# Copy the train images to the train directory
for image in train_images:
    shutil.copy(os.path.join(source_dir, image), os.path.join(train_dir, image))

# Copy the validation images to the validation directory
for image in val_images:
    shutil.copy(os.path.join(source_dir, image), os.path.join(val_dir, image))

# Copy the test images to the test directory
for image in test_images:
    shutil.copy(os.path.join(source_dir, image), os.path.join(test_dir, image))
```
